````python
dia = "ter"
feriado = False

# ERRO Utilizando o simbolo (!) de negação
# (dia == "seg") or !(feriado == False)
# Solução com o ( != )

(dia == "seg") or (feriado != False)

print((dia == "seg") or (feriado != False))
````
